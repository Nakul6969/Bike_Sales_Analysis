```python
import pandas as pd 
```

```python
new = pd.read_excel("bike_selling_dataset.xlsx")

df = new.copy()

print(df.isnull().sum())

print(df.shape)
```

```python
print("Statistics :",df.describe())  # Statistical Summary of the dataset

print("Information :",df.info())  # Information about the dataset   
```

```python
print(df.head())  # First 5 rows of the dataset

print(df.tail())  # Last 5 rows of the dataset
```

```python
# Handling Missing Values in Engine_CC column

df['Engine_CC'] = df.groupby(
    ['Year','Model','Brand']
    )['Engine_CC'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x
)

print(df.isnull().sum())
```

```python
# Handling Missing Values in Mileage_KMPL column

df['Mileage_KMPL'] = df.groupby(
    ['Year','Model','Brand','Engine_CC']
    )['Mileage_KMPL'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x
)

print(df.isnull().sum())
```

```python
# Handling Missing Values in Power_BHP column

df['Power_BHP'] = df.groupby(
    ['Model','Brand','Engine_CC']
    )['Power_BHP'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x
)

print(df.isnull().sum())
```

```python
# since we have 22,000 rows in Dataset , rest of the columns have very less missing values, we can drop those rows

df = df.dropna(subset=['City'])

df = df.dropna(subset=['Seller_Type'])

df = df.dropna(subset=['Condition'])

df = df.dropna(subset=['Kilometers_Driven'])

df = df.dropna(subset=['Price'])
```

```python
print("Shape Of Dataset After Cleaning :", df.shape)

print("Cleaned Values Summary:")
print(df.isnull().sum())
```

```python
# Save cleaned DataFrame to excel file
df.to_csv('cleaned_data.xlsx', index=False)
```
