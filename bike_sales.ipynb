```python
import pandas as pd 
```

```python
new = pd.read_csv("bike_sales_ind.csv")

df = new.copy()

print(df.isnull().sum())

print(df.shape)
```

```python
# we are droping dulicates first and then we will handle missing values

# Remove duplicates
df.drop_duplicates()
```

```python
print("Statistics :",df.describe())  # Statistical Summary of the dataset

print("Information :",df.info())  # Information about the dataset   
```

```python
print(df.head())  # First 5 rows of the dataset

print(df.tail())  # Last 5 rows of the dataset
```

```python
# Handling Missing Values in customer_age column

df["customer_age"] = df["customer_age"].fillna(df["customer_age"].median())

print(df.isnull().sum())
```

```python
# Handling Missing Values in Mileage_KMPL column

df['gender'] = df.groupby(
    ['brand','model','customer_age']
    )['gender'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x
)

print(df.isnull().sum())
```

```python
# Handling Missing Values in Power_BHP column

df['payment_mode'] = df.groupby(
    ['ex_showroom_price','final_price','quantity']
    )['payment_mode'].transform(lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else x
)

print(df.isnull().sum())
```

```python
# since we have 22,000 rows in Dataset , rest of the columns have very less missing values, we can drop those rows

df = df.dropna(subset=['City'])

df = df.dropna(subset=['Seller_Type'])

df = df.dropna(subset=['Condition'])

df = df.dropna(subset=['Kilometers_Driven'])

df = df.dropna(subset=['Price'])
```

```python
print("Shape Of Dataset After Cleaning :", df.shape)

print("Cleaned Values Summary:")
print(df.isnull().sum())
```

```python
# Save cleaned DataFrame to CSV file
df.to_csv('cleaned_Bike_data.csv', index=False)
```
